# 流量趋势图

展现医疗器械网最近半年的流量走势图。

### 导入需要的包

```python
import pylab as plt
plt.rcParams['font.sans-serif']=['SimHei'] #支持中文
```

### 加载数据

```python
label=["1月","2月","3月","4月","5月","6月"] #月份
seouv=[2920 ,2119 ,4008 ,4237 ,4780,4166] #搜索引擎日均UV
dayuv=[3443 ,2581 ,4793 ,5061 ,5567 ,4999] #整体日均UV
```

```python
plt.plot(label,seouv)
plt.plot(label,dayuv)
plt.title("医疗器械网最近半年流量走势图")
plt.xlabel("月份")
plt.ylabel("UV")
plt.legend(["SEO","整体"])
plt.show()
```
